```python
import numpy as np

  # Load libraries
import pandas as pd
from sklearn.tree import DecisionTreeClassifier # Import Decision Tree Classifier
from sklearn.model_selection import train_test_split # Import train_test_split function
from sklearn import metrics #Import scikit-learn metrics module for accuracy calculation
```

```python
wine=pd.read_csv('../input/wine-quality-red-dataset/winequality-red.csv')
```

```python
wine.head()
```

```python
wine.quality.value_counts()
```

```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn import datasets, linear_model
from sklearn.metrics import mean_squared_error, r2_score
```

```python
#split dataset in features and target variable
feature_cols = ['fixed acidity', 'volatile acidity', 'citric acid', 'residual sugar','chlorides','free sulfur dioxide','total sulfur dioxide','density','pH','sulphates','alcohol']
X2 = wine[feature_cols] # Features
y2= wine.quality # Target variable
```

```python
# Split dataset into training set and test set
X2_train, X2_test, y2_train, y2_test = train_test_split(X2, y2, test_size=0.3, random_state=1)
```

```python
# Create Decision Tree classifer object
clf = DecisionTreeClassifier()

# Train Decision Tree Classifer
clf = clf.fit(X2_train,y2_train)
```

```python
#Predict the response for train dataset
y2_pred_train = clf.predict(X2_train)
```

```python
#Predict the response for test dataset
y2_pred = clf.predict(X2_test)
```

```python
# Model Accuracy, how often is the classifier correct?
print("Train Accuracy:",metrics.accuracy_score(y2_train, y2_pred_train))
```

```python
# Model Accuracy, how often is the classifier correct?
print("Test Accuracy:",metrics.accuracy_score(y2_test, y2_pred))
```
